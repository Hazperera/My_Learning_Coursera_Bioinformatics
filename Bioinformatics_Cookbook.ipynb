### Finding the Chloroquine Resistance Transporter (CRT) gene in Plasmodium falciparum on the Nucleotide database

##### Entrez
- provides code to access NCBI over the WWW

##### Handle 
- provide a standard way to deal with information stored in different ways & allow text information to be read incrementally, instead of all at once. A handle is typically a file opened for reading, but could be the output from a command line program, or data downloaded from the internet

##### RetMax 
- the standard search will limit the number of record references to 20, so if  there is more, the  query has to be repeated with an increased maximum limit. hence the default limit is overridden with retmax. The Entrez system provides quite a few sophisticated ways to retrieve large number of results.
Although you now have the IDs of all of the records, you still need to retrieve the records properly.

```python
Entrez.email = 'memyselfni280@.gmail.com'
handle = Entrez.esearch(db='nucleotide', term='CRT[Gene Name] AND "Plasmodium falciparum"[Organism]') #search the nucleotide database for our gene and organism 
rec_list = Entrez.read(handle) #read the result that is returned

if rec_list['RetMax'] < rec_list['Count']:
    handle = Entrez.esearch(db='nucleotide', term='CRT[Gene Name] AND "Plasmodium falciparum"[Organism]', retmax=rec_list['Count'])
    rec_list = Entrez.read(handle)
```

```python
id_list = rec_list['IdList']
hdl = Entrez.efetch(db='nucleotide', id=id_list, rettype='gb')
```

```python
recs = list(SeqIO.parse(hdl, 'gb'))
```
